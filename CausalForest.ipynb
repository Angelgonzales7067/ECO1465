# CAUSAL FOREST

## FEMALE X TYPE1

```python
from econml.dml import CausalForestDML

Y = 'callback'
T = 'female_type1'
controls = ["ba_quality", "language_skills", 
              "exp_highquality", "ma", "certificate",
              'occ_Administrative', 'occ_Biotech and Pharmacy', 'occ_Civil Engineer',
              'occ_Clerical', 'occ_Ecommerce', 'occ_Education',
              'occ_Electrical Engineer', 'occ_Executive Assistant', 'occ_Finance',
              'occ_Food Services Managers', 'occ_Human Resources Payroll',
              'occ_Insurance', 'occ_Maintenance Technician',
              'occ_Marketing and Sales', 'occ_Media and Arts', 'occ_Production',
              'occ_Programmer', 'occ_Retail', 'occ_Social Worker', 'occ_Technology']


# arrays
y = df[Y].values
t = df[T].values
X = df[controls].values

# ----- Causal Forest -----
est1 = CausalForestDML(
    discrete_treatment=True,
    discrete_outcome=True,   
    n_estimators=1000,
    random_state=123,
    inference=True
)

est1.fit(y, t, X=X)
# ----- predicted treatment effect for whole sample -----
df['cate_hat1'] = est1.effect(X)
lb1, ub1 = est1.effect_interval(X)
df['cate_lb1'] = lb1
df['cate_ub1'] = ub1
```

## DISTRIBUTION CATE

```python
import matplotlib.pyplot as plt
import numpy as np

# variables
cate1 = df['cate_hat1'].values
lb1 = df['cate_lb1'].values
ub1 = df['cate_ub1'].values

# lower and upper bounds
lb_line1 = np.mean(lb1)
ub_line1 = np.mean(ub1)

plt.figure(figsize=(8,5))
plt.hist(cate1, bins=50)

# lower bound
plt.axvline(lb_line1, color='black', linestyle='--', linewidth=2,
            label=f"LB mean = {lb_line1:.3f}")

# upper bound
plt.axvline(ub_line1, color='black', linestyle='--', linewidth=2,
            label=f"UB mean = {ub_line1:.3f}")
plt.xlim(-0.3, 0.3)

plt.title("Distribution of Causal Forest CATEs for Female x Type 1")
plt.xlabel("Conditional Average Treatment Effect")
plt.ylabel("Frequency")
plt.legend()
plt.savefig("cate_female_type1.jpg", dpi=300, bbox_inches='tight')
plt.show()
```

# DML

## LINEAR DML

```python
from econml.dml import LinearDML
Y = 'callback'
T = 'female_type1'
controls = ["ba_quality", "language_skills", 
              "exp_highquality", "ma", "certificate",
              'occ_Administrative', 'occ_Biotech and Pharmacy', 'occ_Civil Engineer',
              'occ_Clerical', 'occ_Ecommerce', 'occ_Education',
              'occ_Electrical Engineer', 'occ_Executive Assistant', 'occ_Finance',
              'occ_Food Services Managers', 'occ_Human Resources Payroll',
              'occ_Insurance', 'occ_Maintenance Technician',
              'occ_Marketing and Sales', 'occ_Media and Arts', 'occ_Production',
              'occ_Programmer', 'occ_Retail', 'occ_Social Worker', 'occ_Technology']

# variables
y = df[Y].values
treatments = ['female_type1', 'female_type2', 'female_type3', 'female_type4']
X = df[controls].values

# Create loop for linear DML estimates for all treatments
for i, treat in enumerate(treatments, 1):

    print(f"Running LinearDML for treatment {i}...")

    t = df[treat].values

    est_lin = LinearDML(
        discrete_treatment=True,
        discrete_outcome=True,
        cv=3,
        random_state=123
    )

    est_lin.fit(y, t, X=X)

    # store just the CATEs
    df[f'lin_cate_hat{i}'] = est_lin.effect(X)
```

## NON LINEAR DML

```python
from econml.dml import NonParamDML
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

# variables
y = df[Y].values
X = df[controls].values

treatments = ['female_type1', 'female_type2', 'female_type3', 'female_type4']

# Create loop for Non-Linear DML for all treatments
# The model for Y is a RF and the model for T is a logit, final model is a RF
for i, treat in enumerate(treatments, 1):
    print(f"Running NonParamDML for treatment {i}...")
    
    t = df[treat].values
    
    est_np = NonParamDML(
    model_y=RandomForestClassifier(
        n_estimators=300,
        max_depth=4,
        min_samples_leaf=50,
        random_state=123
    ),
    model_t=LogisticRegression(max_iter=1000), 
    model_final=RandomForestRegressor(
        n_estimators=400,
        max_depth=3,
        min_samples_leaf=50,
        random_state=123
    ),
    discrete_treatment=True,
    discrete_outcome=True,
    cv=3,
    random_state=123
)

    est_np.fit(y,t,X=X)

    # store just the CATEs
    df[f'np_cate_hat{i}'] = est_np.effect(X)
```

```python
# Define cumulative gain function
def cumulative_gain(pred, y, t, min_periods=30, steps=100):
    tmp = pd.DataFrame({'pred':pred, 'y':y, 't':t})
    size = len(tmp)
    tmp = tmp.sort_values('pred', ascending=False)

    rows_list = list(range(min_periods, size, size//steps)) + [size]

    gains = []
    for r in rows_list:
        sub = tmp.head(r)
        num = ((sub.t - sub.t.mean()) * (sub.y - sub.y.mean())).sum()
        den = ((sub.t - sub.t.mean())**2).sum()
        sens = num / den
        gains.append( sens * (r/size) )

    return np.array(gains)

# Define sensitivity function 
def sensitivity(data, y, t):
    return (np.sum((data[t] - data[t].mean())*(data[y] - data[y].mean())) /
            np.sum((data[t] - data[t].mean())**2))
```

```python
# Cumulative Gain plots for all treatments
for i in range(1,5):

    pred_lin = df[f'lin_cate_hat{i}'].values
    pred_np  = df[f'np_cate_hat{i}'].values

    y = df['callback'].values
    t = df[f'female_type{i}'].values

    cg_lin = cumulative_gain(pred_lin, y, t)
    cg_np  = cumulative_gain(pred_np, y, t)

    x = np.linspace(0,1,len(cg_lin))

    # properly select T for the baseline
    baseline = sensitivity(df, 'callback', f'female_type{i}')

    plt.figure(figsize=(12,8))
    plt.plot(x, cg_lin, label='Linear DML')
    plt.plot(x, cg_np,  label='NonParam DML')

    # correct dashed line
    plt.plot([0,1], [0, baseline], '--', color='black', label='Random baseline')

    plt.title(f'Cumulative Gain Curve (Female x Type {i})')
    plt.xlabel('Fraction of population')
    plt.ylabel('Cumulative gain')
    plt.legend()
    plt.savefig(f"cumulative_gain_{i}.jpg", dpi=300,bbox_inches='tight')
    plt.show()
```
